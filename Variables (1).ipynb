# Variables in Python 

****Variables are containers for storing data values****

****Data can be anything like integers, float, complex data,array of numbers, strings etc****
* **We do not need to explicitly define variable type in Python**
* *The type of data stored in a variable decides the type of variable, for example a float data type stored in a variable will automatically convert the variable to float type*

```python
s = 5 #here x is a variable and 5 is a data type
print(s)
type(s)
```

```python
u = 'Tom&Jerry'
print(u)
type(u)
```

```python
p = 3.1415
print(p)
type(p)
```

```python
e = [1, 1.1, 3+7j ,'baby_shark']
print(e)
type(e)
```

```python
r = (1, 1.1, 3+7j, 'dum-duh')
print(r)
type(r)
```

```python
m = 123
m = 'Artificial Intelligence'
print(m)
type(m)
```

### 1) Casting

```python

name  = 190
print(type(name))

a = float(name)
print(type(a))

b = str(name)
print(type(b))

st = '12345a'
print(int(st))
```

### 2) Case-Sensitive

***Python is case sensitive***

```python
n = 'Baby-shark'
N = 7
print(type(n))
print(type(N))
```

# Variable Names

****variable name can either be short name like (x or y), or it could be more descriptive for example (name, place, age, total etc)****

**1) A variable must start with a letter or underscore character**

```python
name = 'Superman'
_place = 'Agriculture University'
print(name)
print(_place)
```

**2) Variable name cannot start with a number**

```python
#07name = 'James Bond'
#print(007name)

7name = 'James Bond'
print(7name)
```

```python
12name = 'James Bond'
print(12name)
```

**We cannot use Hyphens in the variable name**

```python
new-name = 'This is not correct'
print(new-name)
```

**We Cannot Use Spaces between the words when naming a variable**

```python
new name = 'Not Allowed'
print(new name)
```

**3) A variable can contain Alpha-numeric characters and underscores (A-z, 0-9 and _)**

```python
name007 = 'James Bond'
_p_l_ = 'Hollywood'
print(name007)
print(_p_l_)
```

**4) Variable names are case-sensitive meaning (name, Name, NAME, NaMe, nAmE) are all different variables**

```python
name = "Iron Man"
Name = "Captain America"
NAME = "Thor"
NaMe = "Hulk"
nAmE = "Black Widow"
namE = "Spider Man"
print(name)
print(Name)
print(NAME)
print(NaMe)
print(nAmE)
print(namE)
```

## Multi Words Variable Name

**Variable names with multiple words can be difficult to read**
**We can use different techniques to make them more readable**

### 1) Camel Case

**In Camel Case techinque each word in a variable name starts with capital letter except for the first word**

```python
thisNewVariable = 'Hardest choices requires the strongest will'
print(thisNewVariable)
```

## 2) Pascal Case 

**Each word of a variable name Starts With A Capital Letter**

```python
ThisNewVariable = 'Hardest Choices Requires The Strongest Will'
print(ThisNewVariable)
```

## 3) Snake Case

**Each word of a variable name is separated by underscore ( _ )**

```python
this_new_name = 'Thanos'
print(this_new_name)
```

# Many to Many

#### Multiple values to Multiple variables

```python
a, b, c = 'Red', 'Green', 'Blue'
print(a)
print(b)
print(c)
```

# One to Many

#### One Value to Multiple Variables

```python
d = e = f = 'Orange'
print(d)
print(e)
print(f)
```

# Retieve Values from a Collection into Variables 

***Python allows us to extract values from lists and tuples and store them in variables***

```python
f, g, h, i = ['Red', 'Green', 'Yellow', 'Blue']
print(f)
print(g)
print(h)
print(i)
```

```python
j, k, l, m = ('Orange', 'Greed', 'Purple', 'Black')
print(j)
print(k)
print(l)
print(m)
```

****Another way to retireve values from lists and tuples****

```python
Colors = ['Red', 'Green', 'Yellow', 'Blue']
f, g, h, i = Colors
print(f)
print(g)
print(h)
print(i)
```

```python
colors = ('Orange', 'Greed', 'Purple', 'Black')
j, k, l, m = colors
print(j)
print(k)
print(l)
print(m)
```

## More Values and less Variables

****If we have less number of variables and more items in a list or tuple, then we can simply store all the remaining items in a single variable in the form of list****

***We Use Asterisk ( * ) with a variable name in which we want to store multiple values***

```python
colors = ['Red', 'Green', 'Orange', 'Yellow', 'Purple']
x, y, *z = colors 
print(x)
print(y)
print(z)
```

```python
vegies = ('Cauli-flower', 'tomato', 'potato', 'Onion', 'lemon', 1, -3, 3.44, 3+66j)
a, b, c, *d = vegies
print(a)
print(b)
print(c)
print(d)
```
